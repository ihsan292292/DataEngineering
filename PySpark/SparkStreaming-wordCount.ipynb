```python
from pyspark.streaming import StreamingContext
from pyspark.streaming.kafka import KafkaUtils
from pyspark.sql import SparkSession
from pyspark import SparkContext
```

```python
spark = SparkSession.builder.appName("SparkStreaming").getOrCreate()
```

```python
#initilize streaming context with batch interval of 5 seconds
ssc = StreamingContext(spark.sparkContext,5)
```

```python
ssc
```

```python
# create a dstream that will connect to url with localhost as IP and port number as 9990
lines = ssc.socketTextStream("localhost",9990)
words = lines.flatMap(lambda line: line.split(" "))
pairs = words.map(lambda word: (word,1))
wordCount = pairs.reduceByKey(lambda x,y :x+y)
wordCount.pprint()
ssc.start()
ssc.awaitTermination()
```
